# Loch Ness Monster

Plot y = 2 + e**(-bx) * sin(x)

for a domain  x = 0 to 30 taking b = 0.05 and with 100 points.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Set up a domain from 0 to 30 with 100 points
x=np.linspace(0,30,100)
print("Domain Values (x)")
print(x)
print()

# Calculate sin(x)
sinx = np.sin(x)
print("sin of x Values")
print(sinx)
print()

# Calculate e raised to the negative .05 times the value of x
e = math.e
print("Value of e")
print(e)
print()

exponent = (-.05 * x)
print("Exponent Values")
print(exponent)
print()

values = (e**exponent)
print("e Raised to the Exponent Values")
print(values)
print()

y = (2 + (values * sinx))
print("Values of y")
print(y)
print()

print()
print("Graph of y = 2 + e**(-.05x) * (sin(x))")
print()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

Now with b = 0.1

```python
exponent = (-.1 * x)
print("Exponent Values")
print(exponent)
print()

values = (e**exponent)
print("e Raised to the Exponent Values")
print(values)
print()

y = (2 + (values * sinx))
print("Values of y")
print(y)
print()

print()
print("Graph of y = 2 + e**(-.1x) * (sin(x))")
print()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

For both plots, y is a number dependent on x times sine of x plus a constant of 2. I would expect to see a continuous lines with y intercepts of 2 - because our range starts with 0.  With an x value of 0, e**(Anything times 0) times sin(0) = 0 which makes y = 2 in both cases when x is 0.  A normal sine curve's range (y value) is between 1 and -1. In both of these cases, the range of the y variable is determined by the coefficient of sine (e**(negative constant times x)).  I would expect the range to decrease as x increases because e raised to a larger negative power becomes a smaller number to multiply by the value of sine of x.  It would decrease faster in the case of b = .1 than in the case of b = .05 as the product of bx would be larger faster with a b of .1 than a b of .05.

In both examples (b = .05 and b = .1), at both x values selected (4 and 27), the slope is negative (heading in a downward direction).  The graphs confirm this finding.  Additionally, the further away from 0 x becomes, the less steep of a curve exists.  The estimated slopes for changes in x of .1 and .01 between the calculations are different for the same x value as would be expected when estimating.  However, they are very close to each other.  The further away from 0 x is, the closer the two estimates become as the line is becoming less steep, therefore more straight, at that point. 

Plot the function x = [0,1,2,3,4,5], y =[10,9,6,4,6,10]. Estimate the slope between x=1 and x=2 using numerical differentiation; estimate the slope between x=4 and x=5.

```python
import numpy as np
import matplotlib.pyplot as plt

x = [0,1,2,3,4,5]
y =[10,9,6,4,6,10]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
