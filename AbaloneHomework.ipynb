# Abalone Project

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")

filename = "abalone.data.txt"
# x and y coordinates, whitespace-separated
X = np.loadtxt( filename, delimiter=',', usecols=(2,4) )
model = 5.1 * X[:,0] ** 2

plt.scatter(X[:, 0],X[:, 1], marker = 'o', s=150, alpha = 0.5,label='Observed')
plt.scatter(X[:, 0],model,marker='.',color='blue',label=r'Model: W=5.1D$^2$ ')
plt.xlabel('D')
plt.ylabel('W')
plt.legend()

plt.savefig("abalone.png")
```

Compute how well the model fits the observations by computing the SSR. and place this value in a text box on the chart.  The SSR is also know as the loss value.  The closer it is to 0, the better fit your model is for the data.

Sum of square residuals (SSR) = sum of [ ( Yobserved - Yprediction )2 ] which in our example is:

SSR = sum of (X[:, 1] - model)**2


```python
a = 5.1   # Assign the value to a variable so that the same code can be used to see how SSR varies wiht a change in the coefficient of D squared
model = a * X[:,0] ** 2

SSR = np.sum((X[:, 1] - model)**2)
print()
print("The computed SSR for the model W = " + str(a) + "D squared is")
print(SSR)

plt.text(0.05, 2.0, 'SSR = ' + str(SSR))
plt.scatter(X[:, 0],X[:, 1], marker = 'o', s=150, alpha = 0.5,label='Observed')
plt.scatter(X[:, 0],model,marker='.',color='blue',label=r'Model: W = '+str(a)+'D$^2$ ')
plt.xlabel('D')
plt.ylabel('W')
plt.legend()

plt.savefig("abaloneSSR1.png")
```

# Use a Python function to compute the SSR as a function of the observed data and the model parameters.

```python
import numpy as np

filename = "abalone.data.txt"
X = np.loadtxt( filename, delimiter=',', usecols=(2,4) )

x_data = X[:,0]
y_data = X[:,1]
coefficient = 5.1
power = 2

y_model_data = coefficient * x_data ** power
SSR = np.sum(np.square(y_data - y_model_data))
print( SSR )
```
